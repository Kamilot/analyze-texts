```python
def build_text(paragraphs):
    output = ""
    for paragraph in paragraphs.getchildren():
        new_paragraph = True
        for sentence in paragraph.findall("sentence"):
            if not new_paragraph:
                output += " "
            else:
                new_paragraph = False
            output += sentence.find("source").text
        output += "\n"
    return output
    
#print(build_text(paragraphs))
```

```python
def get_prev_word(text, pos):
    current = pos - 1
    while current >= 0 and text[current].isspace() and text[current] != "\n":
        current -= 1
    if current < 0 or text[current] == "\n":
        return None
    word_start = current
    while word_start > 0 and not text[word_start - 1].isspace():
        word_start -= 1
    return text[word_start : current + 1]

def get_next_word(text, pos):
    current = pos + 1
    while current < len(text) and text[current].isspace() and text[current] != "\n":
        current += 1
    if current >= len(text) or text[current] == "\n":
        return None
    word_end = current
    while word_end < len(text) and not text[word_end].isspace():
        word_end += 1
    return text[current : word_end]
```

```python
abbriv_temp = dict()

for text_xml in texts:
    text = build_text(text_xml.find("paragraphs"))
    for i in range(len(text)):
        if (text[i] == "."):
            prev = get_prev_word(text, i)
            next = get_next_word(text, i)
            if next is not None and next[0].islower():
                if prev in abbriv_temp:
                    abbriv_temp[prev] += 1
                else:
                    abbriv_temp[prev] = 1
```

```python
abbriv = {key for key in abbriv_temp if abbriv_temp[key] != 1}
```

```python
def is_term(text, i):
    prev = get_prev_word(text, i)
    next = get_next_word(text, i)
    if next is None:
        return True
    if not next[0].isupper():
        return False
    if text[i] == "." and prev in abbriv:
        return False
    return True
```

```python
punct = [".", "!", "?"]

def find_next_punct(text, i):
    dist = 1
    i += 1
    while i < len(text) and text[i] not in punct and text[i] != '\n':
        i += 1
        dist += 1
    return dist

def find_prev_punct(text, i):
    dist = 1
    i -= 1
    while i >= 0 and text[i] not in punct and text[i] != '\n':
        i -= 1
        dist += 1
    return dist

def mostly_numeric(word):
    cnt = 0
    for c in word:
        cnt += int(c.isnumeric())
    return int(cnt > len(word) / 2)

def mostly_alpha(word):
    cnt = 0
    for c in word:
        cnt += int(c.isalpha())
    return int(cnt > len(word) / 2)
```

```python
def build_feature_vector(text, i):
    features = []
    #features.append(text[:i + 1])
    prev = get_prev_word(text, i)
    next = get_next_word(text, i)
    if (prev is not None):
        features.append(len(prev)) #LSIZE
        features.append(int(prev in abbriv))
        features.append(int(prev[0].isupper()))
        features.append(mostly_numeric(prev))
        features.append(mostly_alpha(prev))
    else:
        features += [0] * 5
    if (next is not None):
        features.append(len(next)) #RSIZE
        features.append(int(next in abbriv))
        features.append(int(next[0].isupper()))
        features.append(mostly_numeric(next))
        features.append(mostly_alpha(next))
        features.append(int(text[i + 1].isspace()))
    else:
        features += [0] * 6
    features.append(find_prev_punct(text, i))
    features.append(find_next_punct(text, i))
    return features
```

```python
def build_text_and_sentence_ends(paragraphs):
    sentence_ends = set()
    output = ""
    for paragraph in paragraphs.getchildren():
        new_paragraph = True
        for sentence in paragraph.findall("sentence"):
            if not new_paragraph:
                output += " "
            else:
                new_paragraph = False
            source = sentence.find("source").text
            sentence_ends.add(len(output) + len(source.lstrip()) - 1)
            output += source
        output += "\n"
    return output, sentence_ends
```

```python
X = []
Y = []

for text_xml in texts:
    text, sentence_ends = build_text_and_sentence_ends(text_xml.find("paragraphs"))
    for i in range(len(text)):
        if (text[i] in punct):
            X.append(build_feature_vector(text, i))
            Y.append(int(i in sentence_ends))
    #break
            
```

```python
from sklearn.svm import SVC
```

```python
classifier = SVC(class_weight="auto")
```

```python
%%time
classifier.fit(X, Y)
```
